```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# colored scateer plots
```

```python
df=pd.read_csv('/content/iris.csv')
df.sample()
```

```python
df['Species']=df['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df.sample(5)
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap='summer')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
```

```python
# plot size

plt.figure(figsize=(15,7))
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap='summer',alpha=1)
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
```

```python
batters=pd.read_csv('/content/batter (1).csv')
batters.head(5)
```

```python
sample_batters=batters.head(100).sample(25,random_state=5)
```

```python
sample_batters
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_batters['avg'],sample_batters['strike_rate'],s=sample_batters['runs'])

for i in range(sample_batters.shape[0]):
  plt.text(sample_batters['avg'].values[i],sample_batters['strike_rate'].values[i],sample_batters['batter'].values[i])
```

```python
x=[1,2,3,4]
y=[5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'green'})
```

```python
# horizontal and vertical lines

plt.figure(figsize=(18,10))
plt.scatter(sample_batters['avg'],sample_batters['strike_rate'],s=sample_batters['runs'])

plt.axhline(130,color='red')
plt.axvline(27.5,color='red')

for i in range(sample_batters.shape[0]):
  plt.text(sample_batters['avg'].values[i],sample_batters['strike_rate'].values[i],sample_batters['batter'].values[i])
```

```python
# Sub plots
# a diff way to plot graphs

plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('something')
plt.xlabel('avg')
plt.ylabel('srike_rate')
```

```python
fig,ax=plt.subplots()
ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('something')
ax.set_xlabel('Avg')
ax.set_ylabel('sr')

fig.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg vs strike rate')
ax[0].set_ylabel('Strike rate')

ax[1].set_title('Avg vs runs')
ax[1].set_ylabel('runs')
```
