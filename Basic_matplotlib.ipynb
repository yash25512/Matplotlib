**Bar chart**

```python
# bivariate analysis, univariate analysis
# numerical vs categorical
# use case - aggregate analysis of groups
```

```python
# simple bar chart
# vertical bar chart
children=[10,20,40,30,20]
colors=['red','green','yellow','blue','pink']

plt.bar(colors,children, color='green')
```

```python
# horizontal bar chart
children=[10,20,40,30,20]
colors=['red','green','yellow','blue','pink']

plt.barh(colors,children, color='green')
```

```python
#multiple bar charts
batsman=pd.read_csv('/batsman_season_record.csv')
batsman
```

```python
plt.bar(batsman['batsman'],batsman['2015'])
```

```python
plt.bar(batsman['batsman'],batsman['2015'],color='red')
plt.bar(batsman['batsman'],batsman['2016'],color='green')
```

```python
plt.bar(np.arange(batsman.shape[0]) -0.2,batsman['2015'],width=0.2)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=0.2)
plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2017'],width=0.2)

plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])
plt.show()
```

```python
# a problem
children=[10,20,40,30,20]
colors=['red red red red red red red red red red','green green green green greeen','yellow','blue','pink']

plt.bar(colors,children)
plt.xticks(rotation='vertical')
```

```python
# stacked bar chart
plt.bar(batsman['batsman'],batsman['2017'],label='2017')
plt.bar(batsman['batsman'],batsman['2016'],bottom=batsman['2017'],label='2016')
plt.bar(batsman['batsman'],batsman['2015'],bottom=(batsman['2016']+batsman['2017']),label='2015')

plt.legend()
plt.show()
```
